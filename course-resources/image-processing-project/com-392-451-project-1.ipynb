```python
!ssh-keygen -t rsa -N "" -C "tmate" -f "/root/.ssh/id_rsa"
```

3. Start the remote shell environment in background.

```python
!tmate -S /tmp/tmate.sock new-session -d
```

4. Get the SSH string to connect to the machine.

```python
!tmate -S /tmp/tmate.sock display -p \'#{tmate_ssh}\'
```
